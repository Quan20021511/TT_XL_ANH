# Các nguổn đầu vào của ảnh cần xử lý và chương trình con show_image và plot_comparison để sử dụng trong những bài tập sau này

```python
# Tải các thư viên cần thiết
from skimage import color, data
import matplotlib.pyplot as plt

# Đọc ảnh sử dụng Matplotlib
image = plt.imread('C:\\Users\\PC\\TT_XULY_ANH\\TUAN_1\\anh_mau.jpg')
image1 = plt.imread('C:\\Users\\PC\\TT_XULY_ANH\\TUAN_1\\anh_chu.jpg')
red_button = plt.imread(r"D:\red.jpg")

# chương trình con show_image()
def show_image(image, title='Image', cmap_type='gray'):
 plt.imshow(image, cmap=cmap_type)
 plt.title(title)
 plt.axis('off')
 plt.show()
    
def plot_comparison(original, filtered, title_filtered):
 fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True, sharey=True)
 ax1.imshow(original, cmap=plt.cm.gray)
 ax1.set_title("original")
 ax1.axis("off")
 ax2.imshow(filtered, cmap=plt.cm.gray)
 ax2.set_title(title_filtered)
 ax2.axis("off")
```

# Bài tập 1.6: Thực hiện phân ngưỡng toàn cục sử dụng phương pháp OTSU trên ảnh:

# Chuyển đổi ảnh màu thành ảnh xám

# Tính toán ngưỡng bằng phương pháp OTSU: sử dụng hàm threshold_otsu() từ thư viện skimage.filters để tính toán ngưỡng phân ngưỡng OTSU cho ảnh.

# Áp dụng kỹ thuật lấy ngưỡng: Dòng code binary = gray_image1 > thresh thực hiện kỹ thuật lấy ngưỡng bằng cách so sánh từng pixel trong ảnh xám gray_image1 với ngưỡng thresh. Kết quả sẽ cho ra một ảnh nhị phân.

```python
#Bài tập 1.6: Phân ngưỡng toàn cục
    
# Khai báo hàm phân ngưỡng OTSU
from skimage.filters import threshold_otsu

gray_image1 = color.rgb2gray(image1)
# Tính toán ngưỡng bằng phương pháp otsu
thresh = threshold_otsu(gray_image1)

# Áp dụng kỹ thuật lấy ngưỡng.
binary = gray_image1 > thresh

# Hiển thị kết quả
show_image(image1, 'Original')
show_image(binary, 'Binary image')
```

# Bài tập 1.7: Thực hiện phân ngưỡng cục bộ (local thresholding) trên ảnh :

# Khai báo sử dụng hàm otsu threshold: sử dụng hàm threshold_local từ thư viện skimage.filters, cho phép chúng ta sử dụng phương pháp Otsu cho phân ngưỡng cục bộ.

# Xác định các ngưỡng cục bộ:  sử dụng hàm threshold_local để xác định các ngưỡng cục bộ trên ảnh xám.

# Ứng dụng ngưỡng cục bộ để xác định ảnh nhị phân: sử dụng các ngưỡng cục bộ đã xác định để áp dụng phân ngưỡng cục bộ trên ảnh xám.

```python
#Bài tập 1.7: Phân ngưỡng cục bộ 

# Khai báo sử dụng hàm otsu threshold
from skimage.filters import threshold_local

# khai báo biến block size
block_size = 35

# xác định các ngưỡng cục bộ với offset=0.05
local_thresh = threshold_local(gray_image1, block_size, offset=0.05)

# Ứng dụng ngưỡng cục bộ để xác định ảnh nhị phân
binary_local = gray_image1 > local_thresh

# hiển thị ảnh gốc và ảnh nhị phân
show_image(image1, 'Original')
show_image(binary_local, 'Local thresholding')

```

# Bài tập 1.8: So Sánh phân ngưỡng cục bộ với phân ngưỡng toàn cục

```python
#Bài tập 1.8: Phân ngưỡng cục bộ khi nền ảnh không rõ ràng và có độ sáng phân bố không đồng đều

# Khai báo sử dụng hàm otsu threshold
from skimage.filters import threshold_local

# khai báo biến block size
block_size = 35

# xác định các ngưỡng cục bộ với offset=0.05
local_thresh = threshold_local(gray_image1, block_size, offset=0.05)

# Ứng dụng ngưỡng cục bộ để xác định ảnh nhị phân
binary_local = gray_image1 > local_thresh

# hiển thị ảnh gốc và ảnh nhị phân
show_image(image1, 'Original')
show_image(binary_local, 'Local thresholding')

# Tính toán giá trị ngưỡng OTSU của ảnh
global_thresh = threshold_otsu(gray_image1)

# Áp dụng kỹ thuật lấy ngưỡng.
binary_global = gray_image1 > global_thresh

# Hiển thị kết quả
show_image(binary_global, 'Global thresholding')
```

# NHẬN XÉT:

* Nhận xét sự giống nhau, khác nhau giữa local_thresh và global_thresh:

Về sự giống nhau: Cả local_thresh và global_thresh đều được sử dụng để phân ngưỡng ảnh xám và tạo ra ảnh nhị phân.
Cả hai phương pháp đều sử dụng giá trị ngưỡng để chia nhỏ hoặc phân loại các điểm ảnh thành hai nhóm: vùng trắng và vùng đen.

Về sự khác nhau: 
Local_thresh:
-Được tính toán dựa trên phương pháp phân ngưỡng cục bộ.
-Sử dụng hàm threshold_local để tính toán giá trị ngưỡng cục bộ cho mỗi khối (block) của ảnh.
-Giá trị ngưỡng được tính dựa trên đặc điểm của từng khối, đảm bảo rằng ngưỡng được thích nghi với các đặc trưng cục bộ của ảnh.
-Local_thresh thích hợp cho các trường hợp khi nền ảnh không rõ ràng và có độ sáng phân bố không đồng đều.

Global_thresh:
-Được tính toán dựa trên phương pháp phân ngưỡng toàn cục.
-Sử dụng hàm threshold_otsu để tính toán giá trị ngưỡng toàn cục cho toàn bộ ảnh.
-Giá trị ngưỡng được tính dựa trên phân bố tổng thể của độ sáng trong ảnh.
-Global_thresh thích hợp cho các trường hợp khi phân bố tổng thể của ảnh có sự phân biệt rõ rệt giữa các vùng.

* Nhận xét sự khác nhau giữa kết quả local_thresh và global_thresh 

Local_thresh:
-Kết quả là mỗi khối nhỏ trong ảnh có một giá trị ngưỡng riêng, dựa trên phân bố độ sáng trong khối đó.
-Vùng trắng và vùng đen trong ảnh nhị phân được tạo ra từ local_thresh có độ sáng và độ tương phản khác nhau theo từng khối, do đó tạo ra một ảnh nhị phân có biến đổi cục bộ.

Global_thresh:
-Kết quả là một giá trị ngưỡng duy nhất cho toàn bộ ảnh, không phụ thuộc vào từng khối.
-Vùng trắng và vùng đen trong ảnh nhị phân được tạo ra từ global_thresh có độ sáng và độ tương phản đồng nhất trên toàn bộ ảnh, không có sự biến đổi cục bộ.

* Giữa local_thresh và global_thresh thì kết quả nào tốt hơn. Vì sao

Xét về bài bài toán: nền ảnh không rõ ràng và có độ sáng phân bố không đồng đều
Thì local_thresh có thể cho kết quả tốt hơn so với global_thresh. Bởi vì
-Local_thresh cho phép xử lý các vùng nền có độ sáng khác nhau trong ảnh. Bằng cách tính toán các giá trị ngưỡng cục bộ cho từng khối nhỏ trong ảnh, nó có thể tách biệt vùng đối tượng và vùng nền một cách tốt hơn trong trường hợp nền ảnh có độ sáng phân bố không đồng đều. Điều này giúp giảm nhiễu và tạo ra một ảnh nhị phân chính xác hơn.
-Khi nền ảnh có sự biến đổi cục bộ, local_thresh có thể giúp tạo ra các vùng nhị phân có độ tương phản khác nhau theo từng khối. Điều này giúp tách biệt các vùng có sự biến đổi cục bộ trong ảnh, giúp cải thiện tính chính xác của quá trình phân đoạn.
-Local_thresh cung cấp một tham số offset để điều chỉnh ngưỡng cục bộ. Bằng cách điều chỉnh tham số này, bạn có thể kiểm soát độ nhạy của phương pháp đối với sự biến đổi cục bộ trong ảnh.
