```python
# khai bo hm local threshold
from skimage.filters import threshold_local
from skimage.filters import threshold_otsu
import matplotlib.pyplot as plt
from skimage import io, color
def show_image(image, title='Image', cmap_type='gray'):
 plt.imshow(image, cmap=cmap_type)
 plt.title(title)
 plt.axis('off')
 plt.show()
image = plt.imread('C:\\Users\\PC\\TT_XLA\\anh_chu.jpg')
gray_image = color.rgb2gray(image)

block_size = 35
# xc nh cc ngng cc b vi offset=10
local_thresh = threshold_local(gray_image, block_size, offset=10)
# ng dng ngng cc b  xc nh nh nh phn
binary_local = gray_image > local_thresh
# hin th nh gc v nh nh phn
show_image(image, 'Original')
show_image(binary_local, 'Local thresholding')


# Tnh ton gi tr ngng OTSU ca nh
global_thresh = threshold_otsu(gray_image)
# p dng k thut ly ngng.
binary_global = gray_image > global_thresh
# Hin th kt qu
show_image(binary_global, 'Global thresholding')
```
