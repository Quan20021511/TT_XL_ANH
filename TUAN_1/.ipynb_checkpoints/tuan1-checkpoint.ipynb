```python
# Tải các thư viên cần thiết
from skimage import color, data
import matplotlib.pyplot as plt

# Đọc ảnh sử dụng Matplotlib
image = plt.imread('C:\\Users\\PC\\TT_XULY_ANH\\TUAN_1\\anh_mau.jpg')
image1 = plt.imread('C:\\Users\\PC\\TT_XULY_ANH\\TUAN_1\\anh_chu.jpg')

# chương trình con show_image()
def show_image(image, title='Image', cmap_type='gray'):
 plt.imshow(image, cmap=cmap_type)
 plt.title(title)
 plt.axis('off')
 plt.show()
```

```python
# Bài tập 1.1: Kích thước ảnh

plt.imshow(image)
image.shape
```

```python
#Bài tập 1.2: Chuyển ảnh RGB thành ảnh xám

# Thư viện
from skimage import io, color

# Chuyển đổi ảnh thành ảnh xám
gray_image = color.rgb2gray(image)
#  Hiển thị ảnh gốc
show_image(image, 'Original RGB image')
#  Hiển thị ảnh xám
show_image(gray_image, 'Grayscale image')
```

```python
#Bài tập 1.3: Lật ảnh

import numpy as np

# Xoay ảnh theo chiều ngược
filpped_image = np.flipud(image)

show_image(image, 'Original RGB image')
show_image(filpped_image, 'Flipped image')
```

```python
#Bài tập 1.4: Lược đồ mức xám

# Trích xuất kênh màu đỏ
red_channel = image[:,:,0]

# Vẽ lược đồ kênh màu đỏ với khoảng màu [0, 256]
plt.hist(red_channel.ravel(), bins=256)

# Đặt tên cho khung hình và hiển thị
plt.title('Red Histogram')
plt.show()
```

```python
#Bài tập 1.5: Phân ngưỡng toàn cục
    
# Khai báo hàm phân ngưỡng OTSU
from skimage.filters import threshold_otsu

gray_image1 = color.rgb2gray(image1)
# Tính toán ngưỡng bằng phương pháp otsu
thresh = threshold_otsu(gray_image1)

# Áp dụng kỹ thuật lấy ngưỡng.
binary = gray_image1 > thresh

# Hiển thị kết quả
show_image(image1, 'Original')
show_image(binary, 'Binary image')
```

```python
#Bài tập 1.6: Phân ngưỡng cục bộ khi nền ảnh không rõ ràng và có độ sáng phân bố không đồng đều

# Khai báo sử dụng hàm otsu threshold
from skimage.filters import threshold_local

# Khai báo sử dụng hàm otsu threshold
block_size = 35

# xác định các ngưỡng cục bộ với offset=0.05
local_thresh = threshold_local(gray_image1, block_size, offset=0.05)

# Ứng dụng ngưỡng cục bộ để xác định ảnh nhị phân
binary_local = gray_image1 > local_thresh

# hiển thị ảnh gốc và ảnh nhị phân
show_image(image1, 'Original')
show_image(binary_local, 'Local thresholding')

# Tính toán giá trị ngưỡng OTSU của ảnh
global_thresh = threshold_otsu(gray_image1)

# Áp dụng kỹ thuật lấy ngưỡng.
binary_global = gray_image1 > global_thresh

# Hiển thị kết quả
show_image(binary_global, 'Global thresholding')
```

```python
#Bài tập 1.7: Ảnh hưởng của việc lựa chọn thông số lấy ngưỡng đối với việc phân ngưỡng ảnh 

# Khai báo sử dụng hàm try_all
from skimage.filters import try_all_threshold

# Sử dụng hàm phân ngưỡng
fig, ax = try_all_threshold(gray_image1, verbose=False)

# Hiển thị kết quả.
plt.show()
```

```python
#Bài tập 1.8: Ứng dụng phân ngưỡng ảnh

# Khai báo sử dụng hàm try_all
from skimage.filters import threshold_otsu,try_all_threshold
# khai báo hàm local threshold 
from skimage.filters import threshold_local

show_image(gray_image1, 'Original')

# Tính toán giá trị ngưỡng OTSU của ảnh
thresh = threshold_otsu(gray_image1)

# Thực hiện phân ngưỡng 
binary_image = (gray_image1 > thresh)

# HIển thị kết quả
show_image(binary_image, 'Binarized image')

#Sử dụng hàm phân ngỡng
fig, ax = try_all_threshold(gray_image1, verbose=False)
plt.show()

# khai báo biến block size 
block_size = 35

# xác định các ngưỡng cục bộ với offset=10
local_thresh = threshold_local(gray_image1, block_size, offset=0.05)

# Ứng dụng ngưỡng cục bộ để xác định ảnh nhị phân
binary_local = (gray_image1 > local_thresh)

# hiển thị ảnh gốc và ảnh nhị phân
show_image(binary_local, 'Local thresholding')
```
